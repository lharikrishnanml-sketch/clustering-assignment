```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
# Load dataset
dataset = pd.read_csv('Mall_Customers.csv')
# Display dataset
dataset
```

```python
 #Select features
X = dataset.iloc[:, 3:5].values
# Display X
X
# DBSCAN Clustering
from sklearn.cluster import DBSCAN
clustering = DBSCAN(eps=10, min_samples=5)
y_clustering = clustering.fit_predict(X)
# Display cluster labels
y_clustering
```

```python
# Add cluster labels to dataset
supervised = dataset
supervised['Cluster_group'] = y_clustering
# Display supervised dataset
supervised
# Save to CSV
supervised.to_csv("cluster_dbscan.csv", index=False)
# Get attributes
dir(clustering)
# Get cluster labels again
y_clustering
# Get column names
supervised.columns
```

```python
# Visualize clusters using seaborn
import seaborn as sns
facet = sns.lmplot(data=supervised, x=supervised.columns[3], y=supervised.columns[4], 
                   hue=supervised.columns[5], fit_reg=False, legend=True, legend_out=True)

plt.title('DBSCAN Clustering')
plt.show()

# Print clustering statistics
n_clusters = len(set(y_clustering)) - (1 if -1 in y_clustering else 0)
n_noise = list(y_clustering).count(-1)
print(f"\nNumber of clusters: {n_clusters}")
print(f"Number of noise points: {n_noise}")

# Note: DBSCAN doesn't have cluster_centers_ attribute
print("\nNote: DBSCAN does not provide cluster centers.")
print("Cluster label -1 represents noise/outlier points.")
```
