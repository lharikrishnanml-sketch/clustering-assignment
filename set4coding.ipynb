```python
import pandas as pd
data={'Name':['Anu','Raj','Kiran','Diya'],'Marks':[76,88,95,65]}
df=pd.DataFrame(data)
result=df[df['Marks']>80]
print(result)
```

```python
import pandas as pd
data={'Name':['Amit','Sneha','Kunal','Riya'],'Department':['HR','IT','HR','IT'],'Salary':[30000,50000,35000,52000]}
df=pd.DataFrame(data)
result=df.groupby('Department')['Salary'].mean()
print(result)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
data={'Name':['Arjun','Leela','Kiran'],'Gender':['Male','Female','Male']}
df=pd.DataFrame(data)
le=LabelEncoder()
df['Gender_encoded']=le.fit_transform(df['Gender'])
print(df)
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
data={'student':['A','B','C'],'score':[55,80,65]}
df=pd.DataFrame(data)
scaler=MinMaxScaler()
df['normalizedscore']=scaler.fit_transform(df['score'])
print(df)
```
