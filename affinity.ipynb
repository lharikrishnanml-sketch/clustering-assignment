```python
# Select features
X = dataset.iloc[:, 3:5].values
# Affinity Propagation Clustering
from sklearn.cluster import AffinityPropagation
clustering = AffinityPropagation(random_state=42)
y_clustering = clustering.fit_predict(X)
# Get cluster labels
y_clustering
```

```python
# Add cluster labels to dataset
supervised = dataset
supervised['Cluster_group'] = y_clustering
# Display dataset with clusters
supervised
# Save to CSV
supervised.to_csv("cluster_affinity_propagation.csv", index=False)
# Get attributes
dir(clustering)
```

```python
# Get cluster centers
centroids = clustering.cluster_centers_
# Display centroids
centroids
# Display cluster labels
y_clustering
# Get column names
supervised.columns
```

```python
# Visualize clusters using seaborn
import seaborn as sns
facet = sns.lmplot(data=supervised, x=supervised.columns[3], y=supervised.columns[4], 
                   hue=supervised.columns[5], fit_reg=False, legend=True, legend_out=True)
plt.title('Affinity Propagation Clustering')
plt.show()
# Print number of clusters found
n_clusters = len(clustering.cluster_centers_indices_)
print(f"\nNumber of clusters found: {n_clusters}")
print(f"Number of iterations: {clustering.n_iter_}")
```
